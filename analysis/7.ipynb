```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 16
plt.rcParams["font.family"] = "Arial"

import matplotlib as mpl

mpl.rcParams["font.family"] = "Arial"
mpl.rcParams["mathtext.fontset"] = "custom"
mpl.rcParams["mathtext.rm"] = "Arial"
mpl.rcParams["mathtext.it"] = "Arial:italic"
mpl.rcParams["mathtext.bf"] = "Arial:bold"


dataset = ["cve","cve24p"]
```

```python
compounds = pd.read_csv('../data/target_molecules.csv', header=None, names=["index","name","smiles"])

import math

def round_sig1(x: float) -> float:
    """有効数字2桁に四捨五入して丸める"""
    if x == 0:
        return 0.0
    order = math.floor(math.log10(abs(x)))
    scale = 10 ** (order)
    return round(x / scale) * scale

def round_sig2(x: float) -> float:
    """有効数字2桁に四捨五入して丸める"""
    if x == 0:
        return 0.0
    order = math.floor(math.log10(abs(x)))
    scale = 10 ** (order - 1)
    return round(x / scale) * scale


for data in {"cve"}:
    print(f"# Results for method: {data}")
    for index, name, smiles in list(compounds.itertuples(index=False))[10:27]:
        # print(f"## Target molecule: {name} (SMILES: {smiles})")
        suc = 0
        time = []
        evals = []
        for seed in range(42,52):
            try:
                df = pd.read_csv(f'../data/2_rediscovery/evomol/substructure_{index}_{name}_{data}_{seed}.csv')
                if df["Score"].iloc[-1] == 1.0:
                    suc += 1
                    time.append(df["TimeElapsed"].iloc[-1]) 
                    evals.append(df["TotalEval"].iloc[-1])
            except FileNotFoundError:
                pass
        print(name)
        print(suc,
              f"{round_sig2(np.mean(evals)) if suc > 0 else None} ± {round_sig2(np.std(evals)) if suc > 1 else None}", # ± {round_sig2(np.std(evals)) 
              f"{round_sig2(np.mean(time)) if suc > 0 else None} ± {round_sig2(np.std(time)) if suc > 1 else None} s")
        if index == 27:
            print(f"{round_sig2(np.median(evals)) if suc > 0 else None}, {round_sig2(np.median(time)) if suc > 0 else None} s")
```
