```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 16
plt.rcParams["font.family"] = "Arial"

import matplotlib as mpl

mpl.rcParams["font.family"] = "Arial"      
mpl.rcParams["mathtext.fontset"] = "custom"   
mpl.rcParams["mathtext.rm"] = "Arial"      
mpl.rcParams["mathtext.it"] = "Arial:italic" 
mpl.rcParams["mathtext.bf"] = "Arial:bold"


title = {"cv": r"$\mathit{Connect}$+$\mathit{Vinyl}$",
         "cve": r"$\mathit{Unit}$",
         "cv24p": r"$\mathit{Connect}$+$\mathit{Vinyl}$+$\mathit{Comp.}$",
         "cve24p": r"$\mathit{Unit}$+$\mathit{Comp.}$"}

color = {"cv": "tab:green",
         "cve": "tab:blue",
         "cv24p": "tab:red",
         "cve24p": "tab:orange"}
```

```python
# population
# load data
dataset = ["cve","cve24p"]
whole_data = dict()
for operations in dataset:
    data = pd.read_csv(f'../data/1_enumeration/{operations}_nodes.csv')
    data = data["Generation"]
    data = data.value_counts().sort_index()
    data = data.reset_index()
    data.columns = ["distance", "count"]
    whole_data[operations] = data

fig, ax = plt.subplots(figsize=(6.5, 5))

for operations, data in whole_data.items():
    ax.plot(data["distance"],
            data["count"],
            marker=("o" if "EZ" in operations else "s"),
            color=color[operations.split("_")[0]],
            label=title[operations.split("_")[0]] + (r" ($\mathit{E/Z}$)" if "EZ" in operations else ""),
            markeredgecolor="black",
            markeredgewidth=0.8,)

ax.set_yscale("log", base=10)

ax.set_xticks(whole_data["cve"]["distance"])
ax.set_xlabel("Number of Construction Steps")
ax.set_ylabel("Counts of Molecules")
ax.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.2)
ax.legend(title="Ruleset", loc="lower right")

plt.tight_layout()
plt.show()
```
